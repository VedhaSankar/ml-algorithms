```python
df = pd.read_csv ("/kaggle/input/linear-regression-dataset/Linear Regression - Sheet1.csv")
df.isnull().sum()
```

```python
df = df.round(2)
```

```python
df.plot.scatter(x="X", y="Y")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
lr = LinearRegression()
```

```python
x_val = df[['X']].values.reshape(-1,1)
y_val = df[['Y']].round(2)
```

```python
lr.fit(x_val, y_val)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.X, df.Y)
```

```python
prediction = lr.predict(x_test.values.reshape(-1,1))
```

```python
plt.plot(x_test, prediction, color = 'b', label = 'Linear Regression')
plt.scatter(x_test, y_test, color = 'g', label = 'Test Data')
plt.legend()
plt.show()
```
