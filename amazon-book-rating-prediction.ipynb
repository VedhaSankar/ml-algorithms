```python
import time
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
gnb = GaussianNB()
```

```python
df.columns
```

```python
X_train, X_test = train_test_split(df, test_size=0.5, random_state=int(time.time()))
```

```python
used_features =[
    'Reviews', 
    'Price', 
    'Year'
]

x_val = X_train[used_features].values
y_val = X_train["User Rating"].astype(int)

gnb.fit(x_val, y_val)
```

```python
y_pred = gnb.predict(X_test[used_features]).reshape(-1,1)
```

```python
gnb.score(x_val, y_pred)
```
